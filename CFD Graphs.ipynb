# Plot Graphs
### Functions: 
**f.PlotWaterLevel(folder,points,Name,Images,Reverse,Number,dots)** <br>
**folder:** Location of run folder *e.g. r"V:\j24130900\OpenFoam\Run\PU_Combined"* <br>
**points:** Name of output file *e.g. "rwe"* <br>
**Name:** Name to call figure if output to png *e.g. "Centreline Water Elevation"* <br>
**Images:** Instruction to export image, *can be "Y" or "N"* <br>
**Reverse:** This call reverses the orientation of the figure, *can be "Reverse" or False* <br>
**Number:** This is the number of vtks to use, *e.g. 40* <br>
**dots:** Whether or not to plot dots, *can be number of dots e.g. 5 or False* <br>
**sensitivity:** Gap between top water level points to count as outliers, *e.g. 0.2m* <br>
df,fig = f.PlotWaterLevel(PMF,"lwe","",Images,False,Number,60,0.2)
